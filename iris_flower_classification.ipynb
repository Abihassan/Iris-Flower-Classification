```python
import pandas as pd
data=pd.read_csv(r"/content/IRIS.csv")
```

```python
data['species']=data['species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
data
```

```python
data_x=data.drop('species',axis=1)
```

```python
data_y=data['species']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,Y_train,Y_test=tts(data_x,data_y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=11)
```

```python
KNN.fit(X_train,Y_train)
```

```python
predicted=KNN.predict(X_test)
```

```python
predicted
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,predicted)*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test,predicted)
```
